#### Question 5
What is the skewness and kurtosis for the fuel quantity burned in two decimal places?

```python
# To get skewness for the fuel quantity burned
fuel_data.skew(axis = 0, skipna = True) 

# Ans = 15.85
```

```python
# To get Kurtosis for the fuel quantity burned
fuel_data.kurtosis(axis = 0, skipna = True)

#Ans = 651.36
```

#### Question 6
Which feature has missing values and what is the total number of missing value and percentage of the missing rows as a factor of the total number of rows in three decimal places? 

(Enter answer in the format: Feature: xxx, Total: xxx, Percent: xxx)

```python
# To check for missing values
tot_missing_values = fuel_data.isnull().sum()
print(tot_missing_values)

total_number_of_rows = len(fuel_data)
print(total_number_of_rows)

#To get the percentage 
missing_values_as_perecentage = ((tot_missing_values)/(total_number_of_rows))*100
print(missing_values_as_perecentage)

# Feature = Fuel Unit 
# Total = 180 
# Percentage = 0.61%
```
